# NLTK Process 
We will try to analyze the "title" column it will consists of basic nlp techniques such as (tokenization, stemming, etc). 
We will pre-process our text data by removing stopwords, non-alphabatical words and non-punctiuation words

```python
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
from nltk.stem import ISRIStemmer
nltk.download('arabic')
nltk.download('punkt')
from nltk import word_tokenize
from nltk.probability import FreqDist
import arabic_reshaper
import string
```

```python
def handle_arabic(input_string):
    reshaped_text = arabic_reshaper.reshape(input_string)
    return get_display(reshaped_text)
```

```python
# Store the title column into single string to analyze it 
arabic_text_data = ' '.join(final_df['title'])
```

```python
# Tokenize the Arabic text data
tokens = word_tokenize(arabic_text_data)
```

```python
# Remove stopwords
arabic_stop_words = set(stopwords.words('arabic'))
tokens = [token for token in tokens if token not in arabic_stop_words]
```

```python
# Remove punctuation and non-word characters
tokens = [token for token in tokens if token not in string.punctuation]
```

```python
# Remove non-alphabetic tokens
tokens = [token for token in tokens if token.isalpha()]
```

```python
word_freq = FreqDist(tokens)
```

```python
'''We will only get most frequent 10 words 
and we can change it to our desired number'''
most_common_words = word_freq.most_common(10)  
print("Most common words:")
for word, frequency in most_common_words:
    print(f"{word}: {frequency}")
```

Looks like a good result so let's start to plot it. First let's convert this data into dataframe for easy plotting

```python
most_common_df = pd.DataFrame(most_common_words, columns=['Word', 'Frequency'])
```

```python
from bidi.algorithm import get_display
most_common_df['Word'] = most_common_df['Word'].apply(handle_arabic)
```

```python
# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(most_common_df['Word'], most_common_df['Frequency'])
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.title('Most Common Words in Title column')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Author column analysis

*   Next we will do analysis on "author" column. By logic the frequency of author indicates the number of articles or works worked on and published




```python
author_frequency = final_df['author'].value_counts()
```

```python
print("Frequency of each author:")
print(author_frequency)
```

```python
defined_authors = []
for i in range(len(author_frequency.index)):
               defined_authors.append(handle_arabic(author_frequency.index[i]))               
```

```python
author_frequency.index = defined_authors
```

```python
# Sorting the top authors
sorted_top_authors = author_frequency.sort_values(ascending=False)[:10]
plt.figure(figsize=(10, 6))
plt.pie(sorted_top_authors, labels=sorted_top_authors.index, autopct='%1.1f%%')
plt.title('Top Authors by Work Published')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.tight_layout()
plt.show()
```

# Topic Column Analysis

```python
topic_distribution = final_df['topic'].value_counts()
```

```python
# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(topic_distribution.index, topic_distribution.values, color='blue')
plt.xlabel('Topic')
plt.ylabel('Number of Articles')
plt.title('Distribution of Articles Across Different Topics (Bar Chart)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(topic_distribution.values, labels=topic_distribution.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Distribution of Articles Across Different Topics (Pie Chart)')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

Impressively, our dataset reveals a remarkable balance, signifying an absence of bias. This characteristic renders it highly suitable for constructing Machine Learning Models in the future. The even distribution of data across topics ensures a fair representation of various themes, enabling robust and unbiased model training. With such balanced data, we can expect our Machine Learning Models to be more accurate and reliable when making predictions or classifications.
